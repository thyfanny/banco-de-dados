>Importando o pandas e o banco de dados

```python
import pandas as pd

banco_dados = pd.read_csv("StarWars.csv", encoding="ISO-8859-1")

banco_dados = banco_dados[pd.notnull(banco_dados['RespondentID'])]
```

>Transformando as colunas com resposta de Sim ou NÃ£o em True e False

```python
yes_no = {
    "Yes": True,
    "No": False
}

banco_dados['Have you seen any of the 6 films in the Star Wars franchise?'] = banco_dados['Have you seen any of the 6 films in the Star Wars franchise?'].map(yes_no)

banco_dados['Do you consider yourself to be a fan of the Star Wars film franchise?'] = banco_dados['Do you consider yourself to be a fan of the Star Wars film franchise?'].map(yes_no)

'Do you consider yourself to be a fan of the Star Trek franchise?'

banco_dados['Do you consider yourself to be a fan of the Star Trek franchise?'] = banco_dados['Do you consider yourself to be a fan of the Star Trek franchise?'].map(yes_no)

```

>Mapeando as colunas de quais filmes foram vistos e renomeando o nome das colunas

```python
import numpy as np
from numpy import nan

seen_episode1 = {
    "Star Wars: Episode I The Phantom Menace": True,
    np.NaN: False
}

seen_episode2 = {
    "Star Wars: Episode II Attack of the Clones": True,
    np.NaN: False
}

seen_episode3 = {
    "Star Wars: Episode III Revenge of the Sith": True,
    np.NaN: False
}

seen_episode4 = {
    "Star Wars: Episode IV A New Hope": True,
    np.NaN: False
}

seen_episode5 = {
    "Star Wars: Episode V The Empire Strikes Back": True,
    np.NaN: False
}

seen_episode6 = {
    "Star Wars: Episode VI Return of the Jedi": True,
    np.NaN: False
}

map_seen_2 = {
    np.NaN: True,
    False: False
}

banco_dados = banco_dados.rename(columns={
    "Which of the following Star Wars films have you seen? Please select all that apply.": "seen_1",
    "Unnamed: 4": "seen_2",
    "Unnamed: 5": "seen_3",
    "Unnamed: 6": "seen_4",
    "Unnamed: 7": "seen_5",
    "Unnamed: 8": "seen_6",
})

banco_dados['seen_1'] = (banco_dados['seen_1'].map(seen_episode1)).map(map_seen_2)
banco_dados['seen_2'] = (banco_dados['seen_2'].map(seen_episode2)).map(map_seen_2)
banco_dados['seen_3'] = (banco_dados['seen_3'].map(seen_episode3)).map(map_seen_2)
banco_dados['seen_4'] = (banco_dados['seen_4'].map(seen_episode4)).map(map_seen_2)
banco_dados['seen_5'] = banco_dados['seen_5'].map(seen_episode5)
banco_dados['seen_6'] = banco_dados['seen_6'].map(seen_episode6)

banco_dados[banco_dados.columns[3:9]]
```

>Renomeando as colunas de ranking e transformando os valores de string para float

```python
banco_dados = banco_dados.rename(columns={
    "Please rank the Star Wars films in order of preference with 1 being your favorite film in the franchise and 6 being your least favorite film.": "ranking_1",
    "Unnamed: 10": "ranking_2",
    "Unnamed: 11": "ranking_3",
    "Unnamed: 12": "ranking_4",
    "Unnamed: 13": "ranking_5",
    "Unnamed: 14": "ranking_6"
})

transforma_float = {
    "1": 6,
    "2": 5,
    "3": 4,
    "4": 3,
    "5": 2,
    "6": 1
    }


banco_dados['ranking_1'] = banco_dados['ranking_1'].map(transforma_float)
banco_dados['ranking_2'] = banco_dados['ranking_2'].map(transforma_float)
banco_dados['ranking_3'] = banco_dados['ranking_3'].map(transforma_float)
banco_dados['ranking_4'] = banco_dados['ranking_4'].map(transforma_float)
banco_dados['ranking_5'] = banco_dados['ranking_5'].map(transforma_float)
banco_dados['ranking_6'] = banco_dados['ranking_6'].map(transforma_float)

banco_dados[banco_dados.columns[9:15]]
```

>Renomeando colunas sobre opiniÃµes acerca dos personagens e trasformando as strings em floats

```python
banco_dados = banco_dados.rename(columns={
    "Please state whether you view the following characters favorably, unfavorably, or are unfamiliar with him/her.": "Han Solo",
    "Unnamed: 16": "Luke Skywalker",
    "Unnamed: 17": "Princesa Leia",
    "Unnamed: 18": "Anakin Skywalker",
    "Unnamed: 19": "Obi Wan Kenobi",
    "Unnamed: 20": "Imperador Palpatine",
    "Unnamed: 21": "Darth Vader",
    "Unnamed: 22": "Lando Calrissan",
    "Unnamed: 23": "Boba Fett",
    "Unnamed: 24": "C-3PO",
    "Unnamed: 25": "R2-D2",
    "Unnamed: 26": "Jar Jar Binks",
    "Unnamed: 27": "Padame Amidala",
    "Unnamed: 28": "Yoda"
})

map_personagens = {
    "Very favorably": 3,
    "Somewhat favorably": 2,
    "Somewhat unfavorably": 1,
    "Neither favorably nor unfavorably (neutral)": 0,
    "Unfamiliar (N/A)": 0,
    "Very unfavorably": -1
}

for personagem in banco_dados[banco_dados.columns[15:29]]:
  banco_dados[personagem] = banco_dados[personagem].map(map_personagens)

banco_dados[banco_dados.columns[15:29]]
```

```python
def Porcentagem(valor, valorTotal):
  return ((100*valor)/valorTotal)

fan = banco_dados[(banco_dados['Do you consider yourself to be a fan of the Star Wars film franchise?'] == True)]
fan_ST_SW = fan[(fan['Do you consider yourself to be a fan of the Star Trek franchise?'] == True)]

n_fan = fan[fan.columns[2]].sum()
n_fan_ST = fan_ST_SW[fan_ST_SW.columns[2]].sum()

print(f'{Porcentagem(n_fan_ST,n_fan):.2f}% das pessoas que disseram ser fÃ£s de Star Wars tmabÃ©m sÃ£o fÃ£s de Star Trek.')
```
